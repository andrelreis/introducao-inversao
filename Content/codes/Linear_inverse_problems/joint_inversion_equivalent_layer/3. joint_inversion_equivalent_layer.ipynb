# Inversão conjunta grav e mag para camada equivalente 

## Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import prism_grav
import prism_mag
import eqlayer
```

## Gerando as coordenadas de observação

```python
Nx = 100
Ny = 50
area = [-1500.,1500.,-1500.,1500.]
shape = (Nx,Ny)
```

```python
x = np.linspace(area[0],area[1],num=Nx)
y = np.linspace(area[2],area[3],num=Ny) 
yc,xc = np.meshgrid(y,x)
```

```python
spacing = y[1] - y[0]
```

```python
spacing
```

```python
voo = -100.
zc = voo*np.ones_like(xc)
```

```python
coordinates = np.array([yc.ravel(),xc.ravel(),zc.ravel()])
```

## Gerando o modelo grav

### Parâmetros do modelo grav

```python
densities = np.array([3000.,3000.])
```

```python
modelo = np.array([[-50.,50., -600.,250.,200.,350.],
                   [-600.,600.,700.,800.,200.,350.]])
```

#### Cálculo da componente gravitacional do modelo

```python
gz = prism_grav.gravitational(coordinates,modelo,densities,field="g_z")
```

### Gerando um conjunto de dados observados com ruído Gaussiano

```python
np.random.seed(seed=40)
std_noise = 0.02*np.max(np.abs(gz))

r = np.random.normal(0.0, std_noise, gz.size)

print ('%.3f mGal' % std_noise)
```

```python
gz_obs = gz + r
```

## Visualização dos dados observados

```python
title_font = 18
bottom_font = 15
plt.close('all')
plt.figure(figsize=(7,7), tight_layout=True)

plt.xlabel('easting (m)', fontsize = title_font)
plt.ylabel('northing (m)', fontsize = title_font)
plt.title('Observed gz (mGal)', fontsize=title_font)
plt.pcolor(yc,xc,gz_obs.reshape(shape),shading='auto',cmap='jet')
plt.tick_params(axis='both', which='major', labelsize=bottom_font)
cb = plt.colorbar(pad=0.01, aspect=40, shrink=1.0)
cb.ax.tick_params(labelsize=bottom_font)

file_name = 'images/observed_gz'
plt.savefig(file_name+'.png',dpi=300)

plt.show()
```

## Gerando o modelo mag

```python
intensidades = np.array([5.,5.])
```

```python
direcoes = np.array([[-60.,-50.],
                     [-60.,-50.]])
```

```python
modelo = np.array([[-50.,50., -600.,250.,200.,350.],
                   [-600.,600.,700.,800.,200.,350.]])
```

### Componentes do campo magnético

```python
bz = prism_mag.magnetics(coordinates,modelo,intensidades,direcoes,field="b_z")
```

```python
bx = prism_mag.magnetics(coordinates,modelo,intensidades,direcoes,field="b_x")
```

```python
by = prism_mag.magnetics(coordinates,modelo,intensidades,direcoes,field="b_y")
```

### Cálculo da Anomalia de Campo Total

```python
I0,D0 = -40.,-30. ## direçao do campo principal
j0x = np.cos(np.deg2rad(I0))*np.cos(np.deg2rad(D0))
j0y = np.cos(np.deg2rad(I0))*np.sin(np.deg2rad(D0))
j0z = np.sin(np.deg2rad(I0))
tfa = j0x*bx + j0y*by + j0z*bz
```

### Gerando um conjunto de dados observados

```python
np.random.seed(seed=40)
std_noise = 0.02*np.max(np.abs(tfa))

r = np.random.normal(0.0, std_noise, tfa.size)

print ('%.3f nT' % std_noise)
```

```python
tfa_obs = tfa + r
```

## Visualização dos dados magnéticos 

```python
title_font = 18
bottom_font = 15
plt.close('all')
plt.figure(figsize=(7,7), tight_layout=True)

plt.xlabel('easting (m)', fontsize = title_font)
plt.ylabel('northing (m)', fontsize = title_font)
plt.title('Observed TFA (nT)', fontsize=title_font)
plt.pcolor(yc,xc,tfa_obs.reshape(shape),shading='auto',cmap='jet')
plt.tick_params(axis='both', which='major', labelsize=bottom_font)
cb = plt.colorbar(pad=0.01, aspect=40, shrink=1.0)
cb.ax.tick_params(labelsize=bottom_font)

file_name = 'images/observed_TFA'
plt.savefig(file_name+'.png',dpi=300)

plt.show()
```

## Resolvendo o problema inverso 

#### Dados observados

```python
d_obs = np.hstack([gz_obs,tfa_obs])
```

```python
d_obs.shape
```

### Gerando a camada equivalente

```python
depth = 150.
zs = depth*np.ones_like(xc)
ys = yc
xs = xc
```

```python
sources = np.array([ys.ravel(),xs.ravel(),zs.ravel()])
```

### Resolvendo a equação de mínimos quadrados regularizado 

```python
I = np.identity(xs.size)
```

```python
I.shape
```

#### Calculando a matriz de sensibilidade grav

```python
G_grav = eqlayer.sensitivity_grav(coordinates,sources)
```

#### Calculando a matriz de sensibilidade grav

```python
main_field = np.array([-40.,-30.])
magnetization = np.array([-60.,-50.])
```

```python
G_mag = eqlayer.sensitivity_mag(coordinates,sources,main_field,magnetization)
```

#### Calculando a matriz de sensibilidade generalizada

```python
G = np.vstack([G_grav,G_mag])
```

```python
G.shape
```

### Resolvendo o sistema linear

```python
mu = 1e-17
```

```python
p = np.linalg.solve(np.dot(G.T,G) + mu*I,np.dot(G.T,d_obs))
```

```python
p_density = 3000.*p
```

```python
p_magnetization = 5.*p
```

## Dados de gravidade preditos

```python
gz_pred = eqlayer.gz(coordinates,sources,p_density)
```

```python
residuals = gz_obs - gz_pred
mean = np.mean(residuals)
std_deviation = np.std(residuals)
```

## Comparando dado observado com os dados gerado pela camada

```python
title_font = 18
bottom_font = 15
plt.close('all')
plt.figure(figsize=(10,10), tight_layout=True)

plt.subplot(2,2,1)
plt.xlabel('easting (m)', fontsize = title_font)
plt.ylabel('northing (m)', fontsize = title_font)
plt.title('Observed gz (mGal)', fontsize=title_font)
plt.pcolor(yc,xc,gz_obs.reshape(shape),shading='auto',cmap='jet')
plt.tick_params(axis='both', which='major', labelsize=bottom_font)
cb = plt.colorbar(pad=0.01, aspect=40, shrink=1.0)
cb.ax.tick_params(labelsize=bottom_font)

plt.subplot(2,2,2)
plt.xlabel('easting (m)', fontsize = title_font)
plt.ylabel('northing (m)', fontsize = title_font)
plt.title('Predicted gz (mGal)', fontsize=title_font)
plt.pcolor(yc,xc,gz_pred.reshape(shape),shading='auto',cmap='jet')
plt.tick_params(axis='both', which='major', labelsize=bottom_font)
cb = plt.colorbar(pad=0.01, aspect=40, shrink=1.0)
cb.ax.tick_params(labelsize=bottom_font)

plt.subplot(2,2,3)
plt.xlabel('easting (m)', fontsize = title_font)
plt.ylabel('northing (m)', fontsize = title_font)
plt.title('Residuals (mGal)', fontsize=title_font)
plt.pcolor(yc,xc,residuals.reshape(shape),shading='auto',cmap='jet')
plt.tick_params(axis='both', which='major', labelsize=bottom_font)
cb = plt.colorbar(pad=0.01, aspect=40, shrink=1.0)
cb.ax.tick_params(labelsize=bottom_font)

plt.subplot(2,2,4)
n, bins, patches = plt.hist(residuals, 50, density=True, facecolor='g', alpha=0.75)
plt.xlabel('Residuals (mu = %.1f | std = %.3f mGal)' % (mean,std_deviation),fontsize = title_font)
plt.ylabel('Frequency',fontsize = title_font)
plt.title('Histogram of residuals',fontsize = title_font)
plt.grid(True)

file_name = 'images/data_fitting'
plt.savefig(file_name+'.png',dpi=300)

plt.show()
```
